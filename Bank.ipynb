```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="1234",
    database="bank_analysis"
)

cursor = conn.cursor()
accounts=pd.read_csv(r"E:\Mock\Bank\Cleaned Data\Acccounts.csv")
customers=pd.read_csv(r"E:\Mock\Bank\Cleaned Data\Customers.csv")
loan=pd.read_csv(r"E:\Mock\Bank\Cleaned Data\Loan.csv")
transactions=pd.read_csv(r"E:\Mock\Bank\Cleaned Data\Transactions.csv")
```

```python
#customer EDA
customers.head()
customers.info()
customers.describe(include="all")
customers.isnull().sum()
customers.duplicated().sum()
```

```python


#Gender distribution
customers["Gender"].value_counts().plot(kind="bar")

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")

for i, v in enumerate(customers["Gender"].value_counts()):
    plt.text(i, v, str(v), ha="center")

plt.show()
#Age group
customers['Age Group'].value_counts().plot(kind='bar')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')
for i,v in enumerate(customers['Age Group'].value_counts()):
  plt.text(i,v,str(v),ha='center')
plt.show()
```

```python
#city distribution
customers['City'].value_counts().plot(kind='bar')
plt.title('City Distribution')
plt.xlabel('City')
plt.ylabel('Number of Customers')
for i,v in enumerate(customers['City'].value_counts()):
  plt.text(i,v,str(v),ha='center')
plt.show()
#Occupation Distribution
customers['Occupation'].value_counts().plot(kind='bar')
plt.title('Occupation Distribution')
plt.xlabel('Occupation')
plt.ylabel('Number of Customers')
for i,v in enumerate(customers['Occupation'].value_counts()):
  plt.text(i,v,str(v),ha='center')
plt.xticks(rotation=45)
plt.show()

#Credit score analysis
customers["Credit_Score"].plot(kind="hist", bins=20)

plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Number of Customers")

plt.show()
#kyc analysis
customers["KYC_Status"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("KYC Status Distribution")
plt.ylabel("")

plt.show()
```

```python
#Accounts Analysis
accounts.head()
accounts.info()
accounts.describe(include="all")
accounts.isnull().sum()
accounts.duplicated().sum()
```

```python


#Account Type Distribution
accounts["Account_Type"].value_counts().plot(kind="bar")

plt.title("Account Type Distribution")
plt.xlabel("Account Type")
plt.ylabel("Number of Accounts")
for i,v in enumerate(accounts["Account_Type"].value_counts()):
    plt.text(i,v,str(v),ha="center")
plt.show()

#Account status
accounts["Account_Status"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("Account Status Distribution")
plt.ylabel("")
plt.show()

#Balance distribution
accounts["Balance"].plot(kind="hist",bins=20)
plt.title("Balance Distribution")
plt.xlabel("Balance")
plt.ylabel("Number of Accounts")
plt.show()

#aggregate function
print("Minimum Balance:", accounts["Balance"].min())
print("Maximum Balance:", accounts["Balance"].max())
print("Average Balance:", accounts["Balance"].mean())

#Top 10
print("Top 10:",accounts.sort_values(by="Balance",ascending=False).head(10))
#bottom 10
print("Bottom 10:",accounts.sort_values(by="Balance",ascending=True).head(10))
```

```python
#Transactions Analysis
transactions.head()
transactions.info()
transactions.describe(include="all")
transactions.isnull().sum()
transactions.duplicated().sum()
```

```python


#Transaction Type Distribution
transactions["Transaction_Type"].value_counts().plot(kind="bar")
plt.title("Transaction Type Distribution")
plt.xlabel("Transaction Type")
plt.ylabel("Number of Transactions")
for i,v in enumerate(transactions["Transaction_Type"].value_counts()):
    plt.text(i,v,str(v),ha="center")
plt.show()
#ransaction Status
transactions["Status"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("Transaction Status Distribution")
plt.ylabel("")
plt.show()

#Channel
transactions["Channel"].value_counts().plot(kind="bar")
plt.title("Transaction Channel Distribution")
plt.xlabel("Channel")
plt.ylabel("Transactions")
for i,v in enumerate(transactions["Channel"].value_counts()):
    plt.text(i,v,str(v),ha="center")
plt.show()
#aggeregate functions
print("Minimum Amount:",transactions["Amount"].min())
print("Maximum Amount:",transactions["Amount"].max())
print("Average Amount:",transactions["Amount"].mean())

```

```python
#EDA
#Loans Analysis
loan.head()
loan.info()
loan.describe(include="all")
loan.isnull().sum()
loan.duplicated().sum()
loan["Interest_Rate"] = loan["Interest_Rate"].str.replace("%","")
loan["Interest_Rate"] = loan["Interest_Rate"].astype(float)
```

```python


#Loan Type Distribution
loan["Loan_Type"].value_counts().plot(kind="bar")
plt.title("Loan Type Distribution")
plt.xlabel("Loan Type")
plt.ylabel("Number of Loans")
for i, v in enumerate(loan["Loan_Type"].value_counts()):
    plt.text(i, v, str(v), ha="center")
plt.show()
#Loan Status
loan["Loan_Status"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Loan Status Distribution")
plt.ylabel("")
plt.show()
#Loan Amount Distribution
loan["Loan_Amount"].plot(kind="hist", bins=20)
plt.title("Loan Amount Distribution")
plt.xlabel("Loan Amount")
plt.ylabel("Number of Loans")
plt.show()
#Interest rate distribution
loan["Interest_Rate"].plot(kind="hist", bins=15)
plt.title("Interest Rate Distribution")
plt.xlabel("Interest Rate")
plt.ylabel("Number of Loans")
plt.show()

#aggregate functions
print("Minimum:",loan["Loan_Amount"].min())
print("Maximum:",loan["Loan_Amount"].max())
print("Average",loan["Loan_Amount"].mean())
```

```python
#Which Account Type has the highest average balance
accounts.groupby("Account_Type")["Balance"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Average Balance by Account Type")
plt.xlabel("Account Type")
plt.ylabel("Average Balance")
plt.show()
#Identify which cities generate the highest total transaction amount.
transactions.groupby("City")["Amount"].sum().sort_values(ascending=False).plot(kind="bar")
plt.title("Total Transaction Amount by City")
plt.xlabel("City")
plt.ylabel("Total Transaction Amount")
plt.xticks(rotation=45)
plt.show()
#Average Loan Amount by Loan Type
avg_loan=loan.groupby("Loan_Type")["Loan_Amount"].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Average loan amount by loan type")
plt.xlabel("Loan type")
plt.ylabel("Average loan amount")
plt.show()
#average transaction by account type
merged_data=pd.merge(accounts,transactions,on="Account_ID")
average_transaction_amounts = merged_data.groupby("Account_Type")["Amount"].mean().sort_values(ascending=False)
ax = average_transaction_amounts.plot(kind="bar")
plt.title("Average transaction amount by account type")
plt.xlabel("Account type")
plt.ylabel("Average transaction amount")
for i, v in enumerate(average_transaction_amounts):
    plt.text(i, v, f"{v:.2f}", ha="center")
plt.show()
#Top 10 Customers by Total Account Balance
merged_data=pd.merge(customers,accounts,on="Customer_ID")
top_customers=merged_data.sort_values(by="Balance",ascending=False).head(10)
top_customers.plot(x="Customer_ID",y="Balance",kind="bar")
plt.title("Top 10 Customers by Total Account Balance")
plt.xlabel("Customer ID")
plt.ylabel("Total Account Balance")
plt.show()
```
